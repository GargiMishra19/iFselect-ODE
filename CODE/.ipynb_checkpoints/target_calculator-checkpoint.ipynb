```python
# Import libraries

import numpy as np
import pandas as pd
from individual_fairness import iFairscore
from normalization import linear_scaling
import target_generator
from pyod.utils.data import evaluate_print
```

```python
# load data

df=pd.read_csv('/Users/Gargi/Desktop/RESEARCH/DATA/InternetAds.csv')
print(df.shape)
df.head()
```

```python
# Separate X(data) and T(target)

T=df['Class'].replace(['inlier', 'outlier'], [0, 1], inplace=False)
X=df.drop(['Class'], axis=1)
```

```python
# Outlier score matrix

# Dataset: 1->IDOS, 2->kNN, 3->LIC, 4->LOF, 5->LoOP, 6->SOD
# Parameter: 0->5, 1->10, 2->20, 3->30, 4->45, 5->60, 6->80, 7->100

Y=pd.read_csv('/Users/Gargi/Desktop/RESEARCH/DATA_SCORE/INTERNETADS_SCORE.csv')
Y
```

```python
Y_norm=linear_scaling(Y)
Y_norm
```

```python
# Psudo ground truth (Target)

t_avg=target_generator.t_average(Y)
t_max=target_generator.t_maximization(Y)
t_maj=target_generator.t_majority_vote(Y)
t_aom=target_generator.t_aom(Y, 8)

evaluate_print("AVG", T, t_avg)
evaluate_print("MAX", T, t_max)
evaluate_print("MAJ", T, t_maj)
evaluate_print("AOM", T, t_aom)

if_avg=iFairscore(t_avg, X)
if_max=iFairscore(t_max, X)
if_maj=iFairscore(t_maj, X)
if_aom=iFairscore(t_aom, X)

print('AVG', if_avg)
print('MAX', if_max)
print('MAJ', if_maj)
print('AOM', if_aom)
```

```python
# Psudo ground truth (Target) - Normalized

t_avg=target_generator.t_average(Y_norm)
t_max=target_generator.t_maximization(Y_norm)
t_maj=target_generator.t_majority_vote(Y_norm)
t_aom=target_generator.t_aom(Y_norm, 8)

evaluate_print("AVG", T, t_avg)
evaluate_print("MAX", T, t_max)
evaluate_print("MAJ", T, t_maj)
evaluate_print("AOM", T, t_aom)

if_avg=iFairscore(t_avg, X)
if_max=iFairscore(t_max, X)
if_maj=iFairscore(t_maj, X)
if_aom=iFairscore(t_aom, X)

print('AVG', if_avg)
print('MAX', if_max)
print('MAJ', if_maj)
print('AOM', if_aom)
```
