```python
import numpy as np
import pandas as pd
from scipy.stats import rankdata
import math
```

```python
# load data
df=pd.read_csv('/Users/Gargi/Desktop/RESEARCH/DATA/St-Mat.csv')
X=df.drop(['Class'], axis=1)
Y=df['Class'].replace(['inlier', 'outlier'], [0, 1], inplace=False)
print(df.shape)
df.head()
```

```python
a=np.array([range(X.shape[0])])
c=a.reshape(X.shape[0],1)
X_idx=np.hstack((c, X.to_numpy()))
```

```python
# Fair Outlier Detector
import math

    
def dist_arr(X):
    summ = 0
    d_arr = np.zeros([X.shape[0], 1])
    for i in range(X.shape[0]):
        for j in range(X.shape[0]):
            summ = summ + math.dist(X[i], X[j])
        d_arr[i] = (summ/(X.shape[0]))
        summ = 0
    idx = np.argmax(d_arr)
    maxi = np.max(d_arr)
    return maxi, idx
    
    
def fit(X):
    dist = np.zeros([X.shape[0], 1])
    for i in range(X.shape[0]):
        maxi, idx = dist_arr(X[:, 1:])
        a_idx = X[idx, 0]
        b=int(a_idx)
        dist[b] = maxi
        X = np.delete(X, (idx), axis=0)
        print(i)
    return dist.ravel()
```

```python
fode = fit(X_idx)
```

```python
from pyod.utils.data import evaluate_print

evaluate_print("FOD", Y, fode)
```

```python
XX = X.to_numpy()
summ=0
dist = np.zeros(XX.shape[0])
r = rankdata(fode)
for i in range(XX.shape[0]):
    for j in range(XX.shape[0]):
        d_diff = math.dist(XX[i], XX[j])
        r_diff = abs(r[i] - r[j])
        d_r = abs(d_diff - r_diff)
        summ = summ + d_r
    dist[i] = (summ/(XX.shape[0]))
    summ=0
ifs=sum(dist)/XX.shape[0]
ifs
```
